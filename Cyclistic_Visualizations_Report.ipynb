# **Cyclistic Case Study: Interactive Data Visualizations**

### **Project Objective:**

> This notebook focuses on creating **interactive visualizations** to explore the behavioral differences between annual members and casual riders. By using the **Plotly** library, we provide a dynamic way to analyze trip durations, weekly patterns, and station popularity.

### **Tools Used:**

* **Python** (Pandas for data handling)
* **Plotly Express** (For interactive charting)



## 1. Data Loading and Preparation

```python

import pandas as pd
import plotly.express as px
import plotly.io as pio


file_path = '/content/drive/MyDrive/Cyclistic_Final_Clean_Data.csv' # Reverting to the previous path as '/content/cleaned_data.csv' was not found.
df = pd.read_csv(file_path, low_memory=False)

# Preview the data
df.head()
```

## 2. Average Trip Duration Analysis
this bar chart compares the average trip duration between casual riders and annual members. It reveals a significant insight: Casual riders tend to cycle for much longer durations per trip compared to members, which suggests they use the service more for leisure or sightseeing rather than quick commutes.

```python

# Calculate average duration per user type
avg_duration = df.groupby('member_casual')['ride_length_mins'].mean().reset_index()

# Create interactive bar chart
fig1 = px.bar(avg_duration,
             x='member_casual',
             y='ride_length_mins',
             title='Average Ride Duration: Casual vs Member',
             labels={'ride_length_mins': 'Average Duration (min)', 'member_casual': 'User Type'},
             color='member_casual',
             color_discrete_sequence=['#e91e63', '#424242'],
             template='plotly_white')

fig1.show()
```

## 2. Weekly Trip Trends: Casual vs. Member

This visualization compares the total number of rides for each day of the week, highlighting how casual riders' activity increases significantly during the weekends

```python

# Ensure days of the week are ordered correctly
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=day_order, ordered=True)

# Group by day and user type to get the count
daily_usage = df.groupby(['day_of_week', 'member_casual']).size().reset_index(name='total_rides')

# Create interactive line chart
fig2 = px.line(daily_usage,
              x='day_of_week',
              y='total_rides',
              color='member_casual',
              title='Total Rides per Day of the Week',
              markers=True,
              labels={'total_rides': 'Number of Rides', 'day_of_week': 'Day of the Week'},
              color_discrete_sequence=['#e91e63', '#424242'],
              template='plotly_white')

fig2.show()
```



### **3. Top 10 Most Popular Start Stations**

> This visualization identifies the top 10 stations where trips originate most frequently. Identifying these high-traffic locations is crucial for the marketing team, as it highlights the best physical locations to launch targeted campaigns and advertisements aimed at converting casual riders into annual members.

```python

# Calculate the top 10 start stations
top_stations = df['start_station_name'].value_counts().head(10).reset_index()
top_stations.columns = ['station_name', 'count']

# Create interactive horizontal bar chart
fig3 = px.bar(top_stations,
             x='count',
             y='station_name',
             orientation='h',
             title='Top 10 Most Popular Start Stations',
             labels={'count': 'Number of Rides', 'station_name': 'Station Name'},
             color_discrete_sequence=['#e91e63'],
             template='plotly_white')

# Improve layout
fig3.update_layout(yaxis={'categoryorder':'total ascending'})
fig3.show()
```



### **4. Most Popular Routes: Top 10 Trips**

> This chart analyzes the most frequent trips by combining start and end stations. It highlights the primary "corridors" used by riders, providing valuable insights into travel patterns and helping to identify the most effective routes for placing physical advertisements or service improvements.



```python

# Create a new column for the route (Start Station to End Station)
df['route'] = df['start_station_name'] + "  to  " + df['end_station_name']

# Calculate the top 10 routes
top_routes = df['route'].value_counts().head(10).reset_index()
top_routes.columns = ['route_name', 'trip_count']

# Create interactive horizontal bar chart
fig4 = px.bar(top_routes,
             x='trip_count',
             y='route_name',
             orientation='h',
             title='Top 10 Most Popular Routes',
             labels={'trip_count': 'Number of Trips', 'route_name': 'Route (Start to End)'},
             color_discrete_sequence=['#e91e63'],
             template='plotly_white')

# Improve layout
fig4.update_layout(yaxis={'categoryorder':'total ascending'})
fig4.show()
```



### **5. Rideable Type Preference: Classic vs. Electric Bikes**

> This visualization compares the usage of different bike types (classic, electric, and docked) between casual riders and members. Analyzing these preferences helps the operations team understand hardware demand and assists the marketing team in tailoring promotions based on the bike types most favored by casual riders.


```python

# Group data by bike type and member type
bike_preference = df.groupby(['rideable_type', 'member_casual']).size().reset_index(name='trip_count')

# Create interactive grouped bar chart
fig5 = px.bar(bike_preference,
             x='rideable_type',
             y='trip_count',
             color='member_casual',
             barmode='group',
             title='Rideable Type Preference: Casual vs Member',
             labels={'trip_count': 'Number of Trips', 'rideable_type': 'Bike Type', 'member_casual': 'User Type'},
             color_discrete_sequence=['#e91e63', '#424242'],
             template='plotly_white')

fig5.show()
```
